```python
# Assuming df is your DataFrame
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
plt.title('Scatter Plot of CGPA vs Resume Score')
plt.show()
```

```python
#iloc[semicolon means all rows,and here we are selecting cols] -> iloc is a function in pandas to split the data
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(x,y)
```

```python
# Coef are actually weights coef = cofficent
p.coef_
```

```python
# intercept is actuaally bias  = intercept is C in ax + by + c =0
p.intercept_
```
